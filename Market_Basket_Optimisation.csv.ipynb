```python
 # Load your InsuranceLosses  dataset
data=pd.read_csv("/content/drive/MyDrive/Imarticus/Stats/Market_Basket_Optimisation.csv")
```

```python
l=[]
for i in range(1,data.shape[0]):
  l.append([str(data.values[i,j])for j in range(0,data.shape[1])])
```

```python
association_rules = apriori(l, min_support=0.0045,min_confidence=0.2,min_lift=3,min_length=2)
association_result = list(association_rules)
```

```python
len(association_result)
```

```python
for i in range(0,len(association_result)):
  print(association_result[i])
```

```python
#Initialize the empty list to store the dictionary representations of the asssociation rules
association_rules_dict = []

#Itrate through the association rules
for record in association_result:
    items = list(record.items)
    support = record.support
    ordered_statistics = []
    for stat in record.ordered_statistics:
        items_base = list(stat.items_base)
        items_add = list(stat.items_add)
        confidence = stat.confidence
        lift = stat.lift
        ordered_statistics.append({
            "items_base" : items_base,
            "items_add" : items_add,
            "confidence" : confidence,
            "lift" : lift
        })

# create a dictionary representations of association rule
rule_dict = {
    "items" : items,
    "support" : support,
    "ordered_statistics" : ordered_statistics
 }

 #Append the rule dictionary to list
association_rules_dict.append(rule_dict)
```

```python
df = pd.DataFrame(columns=["items", "support", "ordered_statistics"])

# Iterate Through the list of dictionaries
for rule in association_rules_dict:
    items = rule["items"]
    support = rule["support"]
    ordered_statistics = rule["ordered_statistics"]

    # Append a new row to the dataframe
    df = df.append({
        "items": items,
        "support": support,
        "ordered_statistics": ordered_statistics
    }, ignore_index=True)

df_sorted = df.sort_values(by="support", ascending=False)

# Display The DataFrame
print(df_sorted)
```
